# Fetching the information from 'ScreenResolution' column
Feature Name ‘ScreenResolution’ contains useful information like whether the laptop is touchscreen or not, whether it has IPS or not, and resolution of the screen. Our Machine cannot interpret that useful information.
By using individual lambda functions for each piece of information, I fetch the information and save it in different columns. Further, by using the split function, I split x resolution and y resolution and find out ppi (Pixel Per Inches).

```python
#Whether the laptop touchscreen or not?
laptop_data['TouchScreen'] = laptop_data['ScreenResolution'].apply(lambda x:1 if 'Touchscreen' in x else 0)
laptop_data.head()
```

```python
#Whether the laptop has IPS or not?
laptop_data['IPS'] = laptop_data['ScreenResolution'].apply(lambda x:1 if 'IPS' in x else 0)
laptop_data.head()
```

```python
#Fetching screen resolution in form of (dddd X dddd)
laptop_data['ScreenResolution']=laptop_data['ScreenResolution'].apply(lambda x: x.split()[-1])
laptop_data.head()
```

```python
#Another method to fetching screen resolution in form of (dddd X dddd)
#laptop_data['ScreenResolution']=laptop_data['ScreenResolution'].str.extract('(\d\d\d\dx\d\d\d\d)')
```

```python
#splitting 2 component of (dddd X dddd). That's known as 'X Resolution' and 'Y Resolution'.
new_df = laptop_data['ScreenResolution'].str.split('x', expand=True)
new_df.head()
```

```python
#Addding 'X_res' and 'Y_res' to our dataframe, laptop_data.
laptop_data['X_res'] = new_df[0]
laptop_data['Y_res'] = new_df[1]
```

```python
laptop_data['X_res']=laptop_data['X_res'].astype('int')
laptop_data['Y_res']=laptop_data['Y_res'].astype('int')
```

# Pixel per Inches (PPI)
Pixels per inch (ppi) are the number of full pixels per inch on a screen, and each pixel on a color display comprises three or more subpixels. Higher resolutions mean that there more pixels per inch (PPI), resulting in more pixel information and creating a high-quality, crisp image. It's calculated as, [sqrt{square(X_resolution) + square(Y_resolution)}]

```python
laptop_data['ppi'] = (((laptop_data['X_res']**2) + (laptop_data['Y_res']**2))**0.5).astype('float')
laptop_data['ppi'] = laptop_data['ppi'].astype('int')
```

```python
laptop_data = laptop_data.drop(columns=['Inches','X_res', 'Y_res', 'ScreenResolution'])
```

# Fetching the information from 'Cpu' column
'Cpu' column contains type of the processor and speed of the processor. By creating a function called process_type, we fetch processor type and creating lambda function we fetch speed of the processor.

# Fetching the information from 'Memory' column
Memory’ contains Storage Type (SSD, HDD, Flash Storage and Hybrid) and Total Storage Capacity. Some laptops have a combination of 2 different storage types. Storage Capacity is in different units GB and TB(1000 GB = 1TB).

```python
laptop_data['Memory'].value_counts()
```

```python
#removed 0 after the decimal point,
#replaced 'GB' with empty string, and 'TB' with '000' (1TB = 1000GB)
#Some records contain 2 different storage types and they are separated by '+' sign.
#Let's split them and save them in new_df2.

laptop_data['Memory'] = laptop_data['Memory'].astype(str).replace('\.0', '')
laptop_data['Memory'] = laptop_data['Memory'].str.replace('GB', '')
laptop_data['Memory'] = laptop_data['Memory'].str.replace('TB', '000')
new_df2 = laptop_data['Memory'].str.split('+', expand=True)
new_df2.head()
```

```python
#we are joing 0th and 1st index column of new_df2 with our original dataframe, creating new column 'para1' and 'para2'
laptop_data['para1'] = new_df2[0]
laptop_data['para2'] = new_df2[1]
laptop_data.head()
```

```python
#We are fetching storage type/s from this piece of code.
#We create four columns Layer1HDD, Layer1SSD, Layer1Hybrid, Layer1Flash_Storage.
#If para1 contains HDD, then 1 will store in Layer1HDD column; otherwise 0. We did same with all four column.
laptop_data['Layer1HDD'] = laptop_data['para1'].apply(lambda x: 1 if 'HDD' in x else 0)
laptop_data['Layer1SSD'] = laptop_data['para1'].apply(lambda x: 1 if 'SSD' in x else 0)
laptop_data['Layer1Hybrid'] = laptop_data['para1'].apply(lambda x: 1 if 'Hybrid' in x else 0)
laptop_data['Layer1Flash_Storage'] = laptop_data['para1'].apply(lambda x: 1 if 'Flash Storage' in x else 0)

#We create another four columns Layer2HDD, Layer2SSD, Layer2Hybrid, Layer2Flash_Storage.
#If para2 contains HDD, then 1 will store in Layer2HDD column; otherwise 0. We did same with all four column.
laptop_data['para2'].fillna("0", inplace = True)
laptop_data['Layer2HDD'] = laptop_data['para2'].apply(lambda x: 1 if 'HDD' in x else 0)
laptop_data['Layer2SSD'] = laptop_data['para2'].apply(lambda x: 1 if 'SSD' in x else 0)
laptop_data['Layer2Hybrid'] = laptop_data['para2'].apply(lambda x: 1 if 'Hybrid' in x else 0)
laptop_data['Layer2Flash_Storage'] = laptop_data['para2'].apply(lambda x: 1 if 'Flash Storage' in x else 0)

laptop_data.sample(5)
```

```python
def mem_type(text):
    if text.split()[-1] == 'SSD':
        return 'SSD'
    elif text.split() [-1] == 'HDD':
        return 'HHD'
    elif text.split() [-1] == 'Storage':
        return 'Flash Storage'
    elif text.split() [-1] == 'Hybrid':
        return 'Hybrid'
    else:
        return ''
```

```python
laptop_data['layer1Memorytype'] = laptop_data['para1'].apply(mem_type)
laptop_data['layer2Memorytype'] = laptop_data['para2'].apply(mem_type)
laptop_data['Memory_type'] = laptop_data['layer1Memorytype'] + ' ' + laptop_data['layer2Memorytype']
laptop_data.drop(columns=['layer1Memorytype', 'layer2Memorytype'], inplace=True)
laptop_data.sample(5)
```

```python
#We removed string from para1 and para2 to convert datatype to float.
laptop_data['para1'] = laptop_data['para1'].str.replace('HDD', '')
laptop_data['para1'] = laptop_data['para1'].str.replace('SSD', '')
laptop_data['para1'] = laptop_data['para1'].str.replace('Hybrid', '')
laptop_data['para1'] = laptop_data['para1'].str.replace('Flash Storage', '')

laptop_data['para2'] = laptop_data['para2'].str.replace('HDD', '')
laptop_data['para2'] = laptop_data['para2'].str.replace('SSD', '')
laptop_data['para2'] = laptop_data['para2'].str.replace('Hybrid', '')
laptop_data['para2'] = laptop_data['para2'].str.replace('Flash Storage', '')

#laptop_data['para1'] = round(laptop_data['para1'], 0)
#laptop_data['para2'] = round(laptop_data['para2'], 0)

laptop_data['para1'] = laptop_data['para1'].astype('float64')
laptop_data['para2'] = laptop_data['para2'].astype('float64')
laptop_data.head()
```

```python
#Now we have eight columns, with 1 and 0 values. Those shows the type of the Storage.
#But we want the actual storage capacity. so we multiplied all 8 columns with para1 and para2.
#So, Now we have Type of storage with their capacity.
laptop_data['HDD'] = laptop_data['para1']*laptop_data['Layer1HDD'] + laptop_data['para2']*laptop_data['Layer2HDD']
laptop_data['SSD'] = laptop_data['para1']*laptop_data['Layer1SSD'] + laptop_data['para2']*laptop_data['Layer2SSD']
laptop_data['Hybrid'] = laptop_data['para1']*laptop_data['Layer1Hybrid'] + laptop_data['para2']*laptop_data['Layer2Hybrid']
laptop_data['Flash_storage'] = laptop_data['para1']*laptop_data['Layer1Flash_Storage'] + laptop_data['para2']*laptop_data['Layer2Flash_Storage']

laptop_data['HDD'] = round(laptop_data['HDD'], 0)
laptop_data['SSD'] = round(laptop_data['SSD'], 0)
laptop_data['Hybrid'] = round(laptop_data['Hybrid'], 0)
laptop_data['Flash_storage'] = round(laptop_data['Flash_storage'], 0)
```

```python
#Drop unnecessory columns
laptop_data.drop(columns=['Memory', 'para1', 'para2', 'Layer1HDD', 'Layer1SSD', 'Layer1Hybrid', 'Layer1Flash_Storage', 'Layer2HDD', 'Layer2SSD', 'Layer2Hybrid', 'Layer2Flash_Storage', 'Memory_type'], inplace=True)
```

```python
#Converting datatype into int
laptop_data['HDD'] = laptop_data['HDD'].astype('int32')
laptop_data['SSD'] = laptop_data['SSD'].astype('int32')
laptop_data['Hybrid'] = laptop_data['Hybrid'].astype('int32')
laptop_data['Flash_storage'] = laptop_data['Flash_storage'].astype('int32')
laptop_data.head()
```

# Fetching the information from 'Gpu' column
A graphics card is an expansion card for your PC that is responsible for rendering images to the display. High-end GPUs are used for gaming, ray tracing, graphics production, and even mining cryptocurrency.The best graphics cards in the current markets are MSI, AMD, and Nvidia. So I considered these cards as good and others as average.

```python
laptop_data['Gpu'].value_counts()
```

```python
def gpu(text):
    if text[0] == 'AMD':
        return 'Good'
    elif text[0] == 'MSI':
        return 'Good'
    elif text[0] == 'Nvidia':
        return 'Good'
    else:
        return 'Average'
```

```python
laptop_data['Gpu'] = laptop_data['Gpu'].apply(gpu)
laptop_data.sample(5)
```

# Fetching the information from 'OpSys' column

```python
laptop_data['OpSys'].value_counts()
```

```python
def operating_system(text):
    if text.split()[0] == 'Windows':
        return 'Windows'
    elif text.split()[0] == 'Linux':
        return 'Linux'
    elif text.split()[0] == 'Mac' and 'macOs':
        return 'Mac'
    elif text.split()[0] == 'Android':
        return 'Android'
    elif text.split()[0] == 'Chrome':
        return 'Chrome'
    else:
        return 'No OS'
```

```python
#Fetching Operatin System
laptop_data['OpSys'] = laptop_data['OpSys'].apply(operating_system)
laptop_data.sample()
```
